#문자열 처리
수집한 데이터의 필요한 문자열을 자르고, 교체하고 추출하는 작업
#string package 사용 : 강력
-str_length() : 문자열 길이
-str_c() : 문자열 연결
-str_sub() : 범위에 해당하는 문자열 생성
-str_split() : 구분자를 기준으로 문자열을 분리하여 부분 생성
-str_replace() : 기존 문자열을 특정 문자열로 변환
-str_extract() : 문자열에서 특정 문자열 패턴의 첫번째 문자열 추출
-str_extract_all() : 모든 문자열 추출
-str_locate():문자열에서 특정 문자열 패턴의 첫번째 위치 추출
-str_locate_all():->모든 문자열 추출


```r
install.packages("stringr")
```

```r
library(stringr)
```

```r
string <- "hongkildong105yoube1002kwanwoo25장비2005"
```

```r
#r 벡터
#a~z까지의 영문자가 연속으로 3개 발견된 문자열 추출
str_extract_all(string, "[a-z]{3}")
```

```r
#a~z까지의 영문자가 연속으로 3~5개 발견된 문자열 추출
str_extract_all(string, "[a-z]{3,5}")
```

```r
#a~z까지의 영문자가 연속으로 1개이상 발견된 문자열 추출
aaa <- str_extract_all(string, "[a-z]{1,}")
aaa
class(aaa)
bbb <- unlist(aaa)
bbb #벡터
#substring안된다..?
```

```r
#숫자만 추출하기
str_extract_all(string, "[0-9]{1,}")#<<<<<
```

```r
aaa <- str_extract_all(string, "[0-9]{1,}")
aaa <- unlist(aaa)
```

```r
aaa<- as.numeric(unlist(aaa))
aaa
```

```r
#한글만 추출하기#ascii코드 table기준? 완성형
aaa <- str_extract_all(string, "[가-힣]{1,}")
aaa <- unlist(aaa)
aaa
```

```r
# 영문자를 제외한 문자 출력하기
str_extract_all(string, "[a-z]{1,}")
str_extract_all(string, "[^a-z]{1,}")
```

```r
#문자열 길이 구하기
str_length(string)
```

```r
#문자열 위치 구하기 #정제 할때 사용? substring못씀
str_locate(string, "장비")
```

```r
string
```

```r
###########################
str_sub(string, 1, 32)#장비 앞
str_sub(string, 1, str_locate(string, "장비")[1])
str_sub(string, 1, str_locate(string, "장비")[1]-1)
###############################
```

```r
#대문자 소문자 변경하기
str_to_upper(string)
str_to_lower(string)
```

```r
#문자열 교체하기 replace
string <- str_replace(string, "hongkildong105", "홍길동35,")
string <- str_replace(string, "yoube1002", "유비45,")
string <- str_replace(string, "kwanwoo25", "관우25,")
string <- str_replace(string, "2005", "20,")
```

```r
#문자열 분리하기 split
str_split(string,",")
unlist(str_split(string,","))
```

```r
#천단위 숫자가 들어간 문자열을 숫자로 변경하기
price <- "123,456,789"
str_replace_all(price,",","")
```
