```r
excel <- read.csv("../Data/excel.csv")
head(excel)
```

```r
#excel에서 q3 컬럼 데이터의 홀수의 합과 짝수의 합 구하기
#출력: 홀수의 합은 __이고 짝수의 합은 __입니다.

q3 <- excel$q3
cat("홀수의 합은",sum(q3[q3%%2==0]),"이고","짝수의 합은", sum(q3[q3%%2!=0]),"입니다.")
```

```r
data <- excel$q3
sumEven<-0
sumOdd<-0

for(i in data){
    if(i%%2==0){
        sumEven <- sumEven +i
    }else{
        sumOdd <- sumOdd +i
    }
}
cat("홀수의 합은",sumOdd,"이고","홀수의 합은",sumEven,"입니다.")
```

```r
cat("홀수의 합은",sum(ifelse(data%%2==1,data,0)),"이고")#함수
cat("짝수의 합은",sum(ifelse(data%%2==0,data,0)),"입니다.")
```

#함수

```r
gugu <- function(dan){

for (i in 1:9){
    cat(dan,"X",i,"=",dan*i,"\n")
}
}
```

```r
gugu(6)
```

```r
for (dan in 2:9) {
  # (1) 바깥 루프: 몇 단인지 제목 출력
  cat(">>>", dan, "단 <<<\n")
  
  # (2) 안쪽 루프: 1부터 9까지 곱하기
  for (i in 1:9) {
    cat(dan, "x", i, "=", dan * i, "\n")
  }
  
  # (3) 한 단이 끝나면 한 줄 띄우기
  cat("\n")
}
```

```r
#2개의 벡터 자료를 함께 출력하기
score <- c(85,95,89)
name <- c("유비","관우","장비")

for (i in 1:length(name)) {
  cat(name[i], ":", score[i], "\n")
}
```

```r
#zip
cat(paste(name,":",score,"\n"),sep = "")
```
