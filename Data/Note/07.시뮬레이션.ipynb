#몬테카를로 시뮬레이션=>빅데이터(데이터가 많아야 정답에 근접함)

```r
#동전의 앞면과 뒷면이 나올 확률이 같다는 것을 증명

coin<-function(n){
    count1 <-0
    count2 <-0
    for(i in 1:n){
        if(runif(1) <= 1/2){
            count <- count1 +1
        }else{
            count2 <- count2+1
        }
    }
    cat("앞면:", count1,"\n")
    cat("뒷면:", count2,"\n")
    cat("-----\n")

}

coin(10) #데이터가 많아지면 근사치에 가까워짐(정답 가까이)
coin(100)
coin(1000)
coin(10000)
coin(100000)
```

```r
#주사위 던지기 시뮬레이션
dice <- function(n){
    diceFace <-rep(0,6)
    cat("Throw",n,"\n")
    for(i in 1:n){
        x<-trunc(runif(1)*6)+1#0하고 1은 포함되지 않아 더해줌
        diceFace[x] <-diceFace[x]+1
    }
    for(i in 1:length(diceFace)){
        cat("No",i,"=>",diceFace[i],"\n")
    }
    cat("-------------\n")
}#코딩테스트 반복문 중요(if문 x)
dice(10)
dice(100)
dice(1000)
dice(10000)
dice(100000)
```
