```r
gugu <- function(range_i,range_j){

for (j in range_j){
    cat(j,"단 \n")


for (i in range_i){
    cat(j,"X",i,"=",j*i,"\n")
}
cat("\n") 
}
}
gugu(1:5, 2:3)
```

```r
#정해진 범위의 구구단 출력 함수
gugu <- function(range_i, range_j) {

  for (j in range_j) {
    cat( j, "단 \n")
    
    
    for (i in range_i) {
      cat(j, "X", i, "=", j * i, "\n")
    }
    cat("\n") 
  }
}

gugu(1:5, 2:3)
```

```r

#결측치(NA) 자료 평균 계산
data <- c(10,20,5,4,40,7,NA,6,3,NA,2,NA)
```

```r
mean(data)
```

```r
#NA의 갯수 구하기
is.na(data)
```

```r
sum(is.na(data))
```

```r
#빈도수
table(is.na(data))
```

```r
#결측치를 무조건 제거하면 통계량을 얻을수 없으며, 데이터 손실로 문제 발생
mean(data, na.rm = T)
```

```r
#선형회귀식 mean근처로 맞춤?
#1)NA를 0으로 대체
data_0<-ifelse(is.na(data),0,data)
#NA 확인 유일한 방법
mean(data_0)
```

```r
#2)NA를 평균으로 대체(계산값, 데이터가 아님)
data_mean<-ifelse(is.na(data),mean(data,na.rm = T),data)
mean(data_mean)
```

```r
#3)NA를 중앙값으로 대체(중앙값은 데이터가 맞음) 중앙값 또는 회귀식 최소데이터 50개 각카테고리별로 딥러닝은 5000개
data_median<-ifelse(is.na(data),median(data,na.rm = T),data)
mean(data_median)####################################
```

```r
#4)NA를 최대값으로 대체
data_max<-ifelse(is.na(data),max(data,na.rm = T),data)
mean(data_max)
```

```r
#5)NA를 최소값으로 대체
data_min<-ifelse(is.na(data),min(data,na.rm = T),data)
mean(data_min)
```
