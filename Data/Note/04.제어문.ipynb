#조건문

```r
x <-10
y <- runif(1) * 100
################
if(x > y){
    print("x > runif()")
    print("x가 큽니다.")
}else if(x<y){
    print("x<runif()")
    print("x가 작습니다.")
}else{
    print("x=runif()")
    print("x와 같습니다.")
}
#################
#[]번째다.->cat
```

#홀짝수 구분

```r
a <- 8
if(a%%2==0){
    cat(a,"는 짝수 입니다.\n")
}else{
    cat(a,"는 홀수 입니다.\n")
}
#스칼라 없음 : 벡터
#리스트 처리 못함
```

```r
a <- c(8, 10, 11)#반복이 있는 삼항연산자
ifelse(a%%2==0,"짝수","홀수")
```

```r
#사용자로 부터 점수를 입력받아 학점으로 변경하여 출력하기
score_input <- readline("점수를 입력하세요:")
score <- as.numeric(score_input)
if(score>=90){
    cat("A")
}else if(score>=80){
    cat("B")
}else if(score>=70){
    cat("C")
}else if(score>=60){
    cat("D")
}else{
    cat("F")
}
```
